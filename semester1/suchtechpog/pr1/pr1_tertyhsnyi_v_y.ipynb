Завдання 3: Створення та вирішення системи лінійних рівнянь

```python
def solve_linear_system(size):
    # Створення матриці A та вектора b з випадковими числами
    A = tf.random.uniform((size, size), minval=1, maxval=10, dtype=tf.float32)
    b = tf.random.uniform((size, 1), minval=1, maxval=10, dtype=tf.float32)
    # Розв'язання системи рівнянь Ax = b
    x = tf.linalg.solve(A, b)
    return A, b, x
```

```python
# Розв'язання системи для різних розмірів
A_5, b_5, x_5 = solve_linear_system(5)
A_10, b_10, x_10 = solve_linear_system(10)
A_100, b_100, x_100 = solve_linear_system(100)
```

Завдання 4: Обчислення градієнтів та побудова графіків функцій

```python
x_vals = tf.linspace(-10.0, 10.0, 500)  # Створення діапазону значень x
```

```python
# Визначення функцій
f1 = x_vals**2  # f(x) = x^2
f2 = tf.sin(x_vals)  # f(x) = sin(x)
f3 = tf.exp(x_vals)  # f(x) = exp(x)
```

```python
# Обчислення функції та її градієнта всередині GradientTape для f(x) = x^2
with tf.GradientTape() as tape:
    tape.watch(x_vals)
    f1 = x_vals**2  # f(x) = x^2
    grad_f1 = tape.gradient(f1, x_vals)  # Похідна f'(x) = 2x

# Обчислення для f(x) = sin(x)
with tf.GradientTape() as tape:
    tape.watch(x_vals)
    f2 = tf.sin(x_vals)  # f(x) = sin(x)
    grad_f2 = tape.gradient(f2, x_vals)  # Похідна f'(x) = cos(x)

# Обчислення для f(x) = exp(x)
with tf.GradientTape() as tape:
    tape.watch(x_vals)
    f3 = tf.exp(x_vals)  # f(x) = exp(x)
    grad_f3 = tape.gradient(f3, x_vals)  # Похідна f'(x) = exp(x)
```

Побудова графіків функцій та їхніх градієнтів

```python
plt.figure()
plt.plot(x_vals, f1, label="f(x) = x^2")  # Графік f(x) = x^2
plt.plot(x_vals, grad_f1, label="f'(x) = 2x")  # Графік похідної f'(x) = 2x
plt.legend()
plt.title("Function and Gradient: x^2")
plt.show()

plt.figure()
plt.plot(x_vals, f2, label="f(x) = sin(x)")  # Графік f(x) = sin(x)
plt.plot(x_vals, grad_f2, label="f'(x) = cos(x)")  # Графік похідної f'(x) = cos(x)
plt.legend()
plt.title("Function and Gradient: sin(x)")
plt.show()

plt.figure()
plt.plot(x_vals, f3, label="f(x) = exp(x)")  # Графік f(x) = exp(x)
plt.plot(x_vals, grad_f3, label="f'(x) = exp(x)")  # Графік похідної f'(x) = exp(x)
plt.legend()
plt.title("Function and Gradient: exp(x)")
plt.show()
```
